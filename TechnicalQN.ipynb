##**2.Python program to print the fibanocci sequence**

```python
user = int(input("How many terms?"))
n1, n2 = 0,1
count = 0
if user <=0:
  print("Please enter a positive integer")
elif user==1:
  print("Fibonacci sequence upto", user, ":")
  print(n1)
else:
  print("Fibonacci sequence:")
  while count < user:
    print(n1)
    N = n1 + n2
    n1 = n2
    n2 = N
    count += 1


```

##**3.Display fibonacci sequence using recursion**

```python
def recur_fibo(n):
  if n <=1:
    return n
  else:
    return(recur_fibo(n-1) + recur_fibo(n-2))
user = int(input("How many number?"))
if user <= 0:
  print("Please enter a positive integer")
else:
  print("fibonacci sequence:")
  for i in range(user):
    print(recur_fibo(i))
```

###**9.Create a list from an iterator object**

```python
class PowTwo:
  def __init__(self, max):
    self.max = max
  def __iter__(self):
    self.num = 0
    return self
  def __next__(self):
    if(self.num>= self.max):
      raise StopIteration
    result = 2** self.num
    self.num +=1
    return result

pow_two = PowTwo(5)
pow_two_iter = iter(pow_two)
print(list(pow_two_iter))
```

##**19.Python program check is a leap year**

```python
year = 2001

if (year % 400==0) and (year % 100 == 0):
  print("{0} is leap year".format(year))

elif (year % 4 ==0) and (year % 100 != 0):
  print("{0} is leap year".format(year))

else:
  print("{0} is not leap year".format(year))
```

##**22.Find the factorial number**

```python
num = int(input("Enter the value: "))
factorial = 1

if num==0:
  print("The number value 0 is the factorial value 0")
elif num<1:
  print("The negative number not be consider factorial value")
elif num>=1:
  for i in range(1,num+1):
    factorial*=i
print("The number {0} of factorial value is {1}".format(num,factorial))
```

##**23.Python program to check the armstrong number**

```python
user = int(input("Enter the value: "))
sum = 0
temp = user

while temp > 0:
  digit = temp % 10
  sum +=digit**3
  temp //=10

if user == sum:
  print(user, "is armstrong number")
else:
  print(user,"is not armstrong number")

```

##**24.Python program to armstrong number in the interval format**

```python
lower =  100
upper = 2000
for num in range(lower, upper+1):
  order = len(str(num))
  sum = 0
  temp = num
  while temp>0:
    digit = temp%10
    sum +=digit**order
    temp//=10
  if num == sum:
    print(num)
```
