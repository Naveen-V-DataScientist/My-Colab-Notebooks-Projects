```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import ElasticNet
import statsmodels.api as sm
```

##**Step1:**

Read the data from pandas and open the dataset

```python
df = pd.read_csv("/content/sample_data/housing_price.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

#**Step2:**

choose the independent variable and dependent variable

```python
x = df.drop(columns = "median_house_value").values
y = df.median_house_value.values
```

#**Step3:**

splitting the data of the train and test

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 12)
```

#**Step4:**
Shape of the train and tes splitting data


```python
print("x_train and x_test dataset shape",x_train.shape,x_test.shape)
print("y_train and y_test dataset shape",y_train.shape,y_test.shape)
```

#**Step5:**

using model is the linear regression model

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

#**Step6:**

the accuracy level of splitting data

```python
regressor.score(x_test,y_test)
```

```python
print("R2 value:",regressor.score(x_test,y_test))
print("\ncoefficient: \n ",regressor.coef_)
print("\nintercept:",regressor.intercept_)
```

#**Step7**

The pridiction of the spliting test data

```python
y_pred = regressor.predict(x_test)
```

```python
pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = pred.head(10)
df1
```

```python
mse = np.mean((df1.Predicted-df1.Actual)**2)
print( "coefficient :",regressor.coef_)
print("\n intercepter :",regressor.intercept_)
print("\n mse :",mse)
print("\n final score",regressor.score(x_test,y_test))
```
