Write  a  Python  program  that  reads  the  Life  Expectancy  Data.csv  (provided  on LMS)  file  into  a  DataFrame,  then  drops  null  values  and performshypothesis testing for the given scenario.H0:u= average life expectancy of Afghanistan is the same as the averageLife expectancy of the world (Null hypothesis)H1:u= average life expectancy of Afghanistan is not the same as the average Life expectancy of the world (Alternate hypothesis)

1.Load the dataset into a data frame

2.Drop all missing values

3.Take   the   mean   of   the   whole   target   variable   and   extract   data   from ‘Afghanistan’4.consider  a  significance  level  of  0.05,  perform  a  T-tail  test  using  the  stats library and calculate P-value and conclude hypothesis testing.

```python
import pandas as pd
import numpy as np
from scipy.stats import t
from scipy.stats import ttest_1samp
```

#**STEP1**
###Load the datasets into a data frame

```python
df = pd.read_csv("/content/sample_data/Life Expectancy Data.csv")
df.head()
```

#**STEP2**
###Drop all missing values

```python
df1 = df.dropna().reset_index()
df1.head()
```

```python
# Checking the any nan value
df1.isnull().sum()
```

#**STEP3**

**NULL HYPOTHESIS(H0):** average life expectancy of Afghanistan is the same as the averageLife expectancy of the world

**ALTERNATE HYPOTHESIS(H1):**  H1 average life expectancy of Afghanistan is not the same as the average Life expectancy of the world



#**STEP4**
Extract the average world of the life expectancy

```python
average_world = df1['Life expectancy '].mean()
average_world
```

#**STEP5**
Extract the average world of the life expectancy and specific country of the afghanisthan

```python
average_afghanistan = df1[df1.Country=="Afghanistan"].sample(12, random_state = 1)
average_afghanistan["Life expectancy "].mean()
```

#**STEP6**  
  **significance level**

a significance level of 0.05 its mean 5% and 95% of the confident level
and obtained from the population of life expectancy and sample of the specific country of the afghanistan data

```python
#significance level
alpha = 0.05
```

#**STEP7**
#####**critical value**
The critical value is probability and degree of freedom calculate to find
#####**p-value, t-statistic**
The importing the library to using and the find

```python
p = 1-alpha/2
df1 = 11
critical_value = t.ppf(p, df1)
print("The critical value of 95% confidence level and with 11 DOF(degrees of freedom):",critical_value)

t_statistic, p_value = ttest_1samp(average_afghanistan["Life expectancy "], average_world)
print('The t-statistic value is',t_statistic)
print('The p-value is',p_value)
```

#**STEP8**
Compare the alpha value(0.05) and p-value using to calculates the reject or not rejecting the null hypothesis

```python
if p_value < alpha:
    print('We  reject the null hypothesis.\nThe average Life expectancy of afghanistan is not same as  average Life expectancy of world.')
else:
    print('We fail to reject  the null hypothesis.\nThe average Life expectancy of afghanistan is  same as average Life expectancy of world.')
```
