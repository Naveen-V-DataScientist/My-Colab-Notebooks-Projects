```python
Qimport cv2
import imutils
from google.colab.patches import cv2_imshow
```

```python
alg =  "/content/sample_data/haarcascade_frontalface_default.xml.2"
har_cascade = cv2.CascadeClassifier(alg)
```

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml -P /content/sample_data/
```

```python
source_data = "/content/sample_data/VID-20220623-WA0030.mp4"
cam = cv2.VideoCapture(source_data)
while True:
  _,img = cam.read()
  if not _: # Check if the frame was read successfully
    print("Failed to read frame from video file. Exiting...")
    break
  grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  faces = har_cascade.detectMultiScale(grayImg, 1.3, 4)
  for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
  cv2_imshow(img)
  key = cv2.waitKey(10)
  if key == 27:
    break
cam.release()
cv2.destroyAllWindows()
```
