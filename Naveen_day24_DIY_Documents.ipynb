##Q1. Problem Statement: Conditional Probability

Load the “kerala.csv”data into a DataFrame and perform the following tasks:

2.June and July are the peak months of rainfall. Consider that if it rains more than  500mm,  then the chances  of  flood  become  more;  create  a  Dataframe with  columns –“YEAR”,  “JUN_GT_500” (Contains  a  boolean  value  to  show whether  it  rained  more  than  500  mm  in June), “JUL_GT_500” (Contains  a boolean  value  to  show  whether  it  rained  more  than  500  mm  in July),and “FLOODS” (Contains a boolean value to show whether it flooded that year)

3.Calculate the probability of flood given it rained more than 500 mm in June (P(A|B))

4.Calculate the probability of rain more than 500 mm in June, given it flooded that year (P(B|A))

5.Probability of flood given it rained more than 500 mm in July

6.Probability  of  rain of more  than  500  mm  in  July  given  it  flooded  that  year (P(B|A))


```python
import pandas as pd
df = pd.read_csv("/content/sample_data/kerala.csv")
df.head()
```

```python
df["JUN_GT_500"] = (df["JUN"] > 500 ).astype("int")
df["JUL_GT_500"] = (df["JUL"] > 500 ).astype('int')
df["FLOODS"] = df['FLOODS'].map({'YES': 1, 'NO':0})
```

```python
df2 = df.loc[:, ["YEAR", "JUN_GT_500","JUL_GT_500","FLOODS"]]
df2["COUNT"] = 1

df2.head()
```

```python
DF3 = pd.crosstab(df2["FLOODS"], df2["JUL_GT_500"])
print(DF3)
```

```python
prob_floods = (6 + 54) / (6 +54 + 19 + 39)
prob_Jun = (39 + 54) / (6 + 54 + 19 + 39)
prob_floods_inter_jun = 54/(6 + 54 + 19 +39)
print("P(Flood): ", prob_floods)
print("P(June): ", prob_Jun)
print("P(Flood AND June): ",prob_floods_inter_jun)
```

```python
# Now calculate probabilitity of flood given it rained more than 500 mm in June (P(A|B))
Proba_Floods_Jun = prob_floods_inter_jun / prob_Jun
print("Probailitity of flood given it rained more than 500 mm in June (P(A|B)): ")
print("P(Flood|June): ", Proba_Floods_Jun)
```

```python
Proba_Jun_Floods = (Proba_Floods_Jun * prob_Jun) / prob_floods
print("Probability of rain more than 500 mm in June given it flooded that year (P(B|A)): ")
print("P(June|Flood): ",Proba_Jun_Floods)
```
