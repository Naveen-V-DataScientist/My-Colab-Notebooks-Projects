#**STEP 1**
Importing necessary libraries

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , classification_report,accuracy_score
from sklearn.metrics import roc_curve , auc
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import sys
````

#**STEP 2**
Load the data

````python
df=pd.read_csv("/content/sample_data/Heart_Disease.csv")
df.head()
````

#**STEP 3**
Describe the dataset in the seven points statistical representation

````python
df.describe()
````

#**STEP 4**
Check the any missing value in the dataset

````python
df.isnull().sum()
````

#**STEP 5**
Fill the missing values


````python
df = df.fillna(df.mean())
df.isnull().sum()
````

#**STEP 6**
Splitting the train and test variable for the input data and ouptput data


````python
X = df.drop("TenYearCHD",axis=1)
y= df["TenYearCHD"]
print("Columns in X :",X.columns)
print("y :",y)
print("shape of X:",X.shape)
print("shape of y:",y.shape[0])
````

#**STEP 7**
Splitting the data into train and test . evaluating the model

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
````

````python
print("After splitting the data-")
print("size of input train data :", sys.getsizeof(X_train))
print("sizeof input test data :", sys.getsizeof(X_test))
print("size of output train data :", sys.getsizeof(y_train))
print("size of output test data :", sys.getsizeof(y_test))
````

#**STEP 8**
Using the StandardScalar() function


```
object = StandardScaler()
object.fit_transform(data)
```



````python
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train=pd.DataFrame(X_train,columns=X.columns)
X_test=pd.DataFrame(X_test,columns=X.columns)
````

#**STEP 9**
Using the model logistic regression to train the data and testing data prediction

````python
model=LogisticRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
````

#**STEP 10**
print the classification

````python
CFR = classification_report(y_test,y_pred)
print(CFR)
````

#**STEP 11**
Print the confusion matrix

````python
conf_mat=confusion_matrix(y_test,y_pred)
print("Confusion matrix is \n",conf_mat)
plt.figure(figsize=(12,9))
plt.title("Logistic Regression Confusion Matrix")
sns.heatmap(conf_mat, square=True,cmap="BuPu",annot=True,fmt='d')
plt.xlabel('true label')
plt.ylabel('predicted label')
````

#**STEP 12**
Find the accuracy value

````python
print("accuracy score : ",accuracy_score(y_test,y_pred))
print("accuracy:", round(100*accuracy_score(y_test,y_pred)),"%")
````
