#**Concatenate**

Concatenate is used for two matrix or string etc...

join the 2 variables, list, array

```python
a = np.array([[1,3],[3,4]])
b = np.array([[2,3],[6,7]])
print("Array1: ")
print(a)
print("Array2: ")
print(b)
con = np.concatenate((a,b), axis = 1)
print("Concatenate: ")
print(con)
```

#**Horizantal stack, Vertical stack**


```python
a = np.array((91,2,4))
b = np.array((3,4,6))
c = np.hstack((a,b))
d = np.vstack((a,b))
print(c)
print(d)
```

#**Splitting arrays**

1.horizantal split
2.vertical split

```python
a = np.arange(16).reshape((4,4))
print(a)

[b,c] = np.split(a,2, axis = 1)
print(b)
print(c)
```

```python
arr = np.hsplit(a,2)
arr
```

```python
arr1 = np.vsplit(a,2)
print(arr1)
```

#**File handling**
file handling is nothing but used for save and load. It is used for open the file and read or write.

```python
x = [[1,2,3],[4,5,6],[7,8,9]]
a = np.array(x)
arr = np.save("file.npy",a)

arr1 = np.savetxt("fil.txt",a)
b = np.genfromtxt("fil.txt",arr1)
c = np.load("file.npy",arr)
print(b)
print(c)
```
