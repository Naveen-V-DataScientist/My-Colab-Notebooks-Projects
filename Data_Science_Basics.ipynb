**The basics of the merge and concatination**

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
x = np.random.randint(0, 100, (4,5))
x
```

```python
df = pd.DataFrame(x, index = list("abcd"), columns = list('vwxyz'))
df
```

```python
np.random.seed(0)
y = np.random.randint(0,100,(4,3))
df1 = pd.DataFrame(y, columns = list("pqr"), index = list("abcd"))
```

```python
df1
```

```python
new_data = pd.concat([df, df1], axis = 1)
new_data
#df3 = pd.merge(df1, df, on = "pat", how = "right" or 'left' or 'inner' or "outer" )
```
