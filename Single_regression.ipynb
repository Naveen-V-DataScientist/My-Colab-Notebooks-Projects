```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/content/sample_data/Advertising.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize = (10, 15))
plt.scatter(df['TV'], df['sales'], color = 'black')
plt.xlabel("Tv ads")
plt.ylabel("Sales")
plt.title("Tv ads vs sales")
plt.show()
```

```python
#x = df.drop(["sales", "Unnamed: 0"], axis = 1)
X = df["TV"].values.reshape(-1,1)
X
```

```python
y = df["sales"].values.reshape(-1,1)
y
```

```python
from sklearn.linear_model import  LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.score(X,y)
```

```python
y_pred = reg.predict(X)
plt.figure()
plt.scatter(df['TV'],df['sales'],c='yellow')
plt.plot(df['TV'],y_pred,c='blue',linewidth=2)
plt.xlabel("TV ads")
plt.ylabel("Sales")
plt.show()
```

**Plot the "TV ads" vs "Sales"**

```python
print(reg.coef_)
print(reg.intercept_)
```
