```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Categories and their possible sources
expense_sources_map = {
    'Food': ['Supermarket', 'Restaurant', 'Grocery Store'],
    'Transport': ['Bus', 'Auto', 'Cab', 'Train'],
    'Shopping': ['Online Store', 'Local Market', 'Mall'],
    'Rent': ['Landlord', 'Apartment Rent'],
    'Utilities': ['Electricity Board', 'Water Supply', 'Gas Company'],
    'Entertainment': ['Cinema', 'Streaming Service', 'Concert'],
    'Medical': ['Pharmacy', 'Hospital', 'Clinic']
}

income_sources = ['Salary', 'Freelance', 'Bonus', 'Investment Return']
payment_modes = ['Cash', 'UPI', 'Credit Card', 'Debit Card', 'Bank Transfer']

# Generate 1000 records starting from a specific date
start_date = datetime(2024, 1, 1)
data = []

for _ in range(1000):
    # Random date between Jan 1, 2024 and Apr 10, 2025
    date = start_date + timedelta(days=random.randint(0, 465))
    month = date.strftime('%B')
    year = date.year

    if random.random() < 0.8:  # 80% chances it's an expense
        category = random.choice(list(expense_sources_map.keys()))
        amount = round(random.uniform(50, 5000), 2)
        payment = random.choice(payment_modes)
        expense_source = random.choice(expense_sources_map[category])
        data.append({
            'Date': date.strftime('%Y-%m-%d'),
            'Category': category,
            'Description': f'{category} expense',
            'Amount': amount,
            'Type': 'Expense',
            'Month': month,
            'Year': year,
            'Payment Mode': payment,
            'Expense Source': expense_source
        })
    else:  # 20% income entries
        source = random.choice(income_sources)
        amount = round(random.uniform(15000, 60000), 2)
        data.append({
            'Date': date.strftime('%Y-%m-%d'),
            'Category': 'Income',
            'Description': f'{source} received',
            'Amount': amount,
            'Type': 'Income',
            'Month': month,
            'Year': year,
            'Payment Mode': 'Bank Transfer',
            'Expense Source': ''
        })

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
file_path = "daily_expense_income_dataset.csv"
df.to_csv(file_path, index=False)

print(f"✅ Dataset generated and saved as '{file_path}'")
```
