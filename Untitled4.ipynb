```python
df = pd.read_csv("/content/sample_data/cereal.csv")
```

```python
mfr={'A':'American Home Food Products','G':'General Mills','K':'Kelloggs',
       'N':'Nabisco','P':'Post','Q':'Quaker Oats','R':'Ralston Purina'}
df2=df['mfr'] = df['mfr'].replace(mfr)
display(df2)
```

```python
sns.scatterplot(x='rating',y='calories',data=df,hue='mfr')
sns.lineplot(x='rating',y='calories',data=df,hue='mfr')
plt.grid()
plt.show()
```
