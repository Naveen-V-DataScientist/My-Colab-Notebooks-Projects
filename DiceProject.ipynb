```python
import random

print("ğŸ¯ Welcome to the Six Dice Game ğŸ¯")
print("----------------------------------")

player1 = input("Enter Player 1 Name: ")
player2 = input("Enter Player 2 Name: ")

# Initialize totals
player1_total = 0
player2_total = 0

# Roll 6 dice for each player
print(f"\n{player1}'s turn:")
for i in range(1, 7):
    roll = random.randint(1, 6)
    print(f"Dice {i}: {roll}")
    player1_total += roll

print(f"Total score for {player1}: {player1_total}")

print(f"\n{player2}'s turn:")
for i in range(1, 7):
    roll = random.randint(1, 6)
    print(f"Dice {i}: {roll}")
    player2_total += roll

print(f"Total score for {player2}: {player2_total}")

# Determine the winner
print("\nğŸ Final Result ğŸ")
if player1_total > player2_total:
    print(f"ğŸ† {player1} wins!")
elif player2_total > player1_total:
    print(f"ğŸ† {player2} wins!")
else:
    print("ğŸ¤ It's a tie!")
```
