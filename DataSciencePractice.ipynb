#**Numpy concept**

```python
import numpy as np
x = [[0, 1, 2],[3, 4, 5], [6, 7, 8]]
arr = np.array(x)
print(arr)
type(arr)
```

##**Zeros  & Ones function**

```python
arr1 = np.zeros((5,5))
print(arr1)
arr2 = np.ones((3,3))
print(arr2)
```

##**Copy, Arange, linspace, lipher, eye, dot**

```python
#Copy
a = np.array([1,2,3,4,5])
c = a.copy()
print(c)
del(c)
```

```python
#Arange
a = np.arange(1,31,2)
print(a, "\n")
#Linspace
c = np.linspace(2,4)
print(c)
type(c)
```

```python
#Mod()
arr = np.mod(9,4)
print(arr)
arr1 = np.eye(2,3)
print("eye:",arr1,"\n")
arr2 = np.char.capitalize("PYTHON")
arr_2 = np.char.lower(arr2)
arr_3 = np.char.upper(arr_2)
print(arr_2)
print(arr_3)
print("Character capitalize:",arr2, "\n")
arr3 = np.char.add("python", " is programming language")
print("concatenation:",arr3,"\n")
arr4 = np.char.multiply("Naveen ",3)
print("Multiply:",arr4, "\n")
arr5 = np.char.mod("Naveen", 3)
print("modulation:", arr5,"\n")

arr6 = np.char.center("Naveen", 2,fillchar ="e")
print("Center function:",arr6, "\n")
arr7 = np.char.join("naveen", "123")
print("join:",arr7, "\n")
arr8 = np.char.ljust("naveen",width = 4)
print("ljust:", arr8, "\n")
arr9 = np.char.swapcase("WhATEVER FINE")
print("Swapcase:",arr9,"\n")
arr10 = np.char.lstrip(np.array(["naveen","dinesh","umamaheshwari","vellingiri","avignesh"]),"n")
print("lstrip:", arr10, "\n")
```

```python
arr11 = np.char.partition("naveen",["n",'a',"v","e","e","n"])
print("Partition:", arr11, "\n")
c = np.array(["A success is kee","kee is another word is lossing"])
arr12 = np.char.replace(c,"kee", "key")
print(c)
print("replace:",arr12, '\n')
arr13 = np.char.ljust("naveen",width = 7,fillchar = "n")
print("rjust:", arr13, "\n")
arr14 = np.char.partition("n","f")
print("rpartition:", arr14, "\n")
a = np.array("naveen is bad boy")
arr15 = np.char.rsplit(a,"n" )
print("splitlines:", arr15, "\n")
```
